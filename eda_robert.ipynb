```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the built-in tips dataset
tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.histplot(tips["total_bill"], kde=True, bins=20)
plt.title("Distribution of Total Bill")
plt.xlabel("Total Bill ($)")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.histplot(tips["tip"], kde=True, bins=20)
plt.title("Distribution of Tip Amount")
plt.xlabel("Tip ($)")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip")
plt.title("Tip vs. Total Bill")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
plt.show()
```

```python
sns.boxplot(data=tips, x="sex", y="tip")
plt.title("Tip Distribution by Sex")
plt.xlabel("Sex")
plt.ylabel("Tip ($)")
plt.show()
```

```python
sns.violinplot(data=tips, x="day", y="tip")
plt.title("Tip Distribution by Day")
plt.xlabel("Day of Week")
plt.ylabel("Tip ($)")
plt.show()
```

```python
sns.heatmap(tips.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
avg_tip_by_day = tips.groupby("day")["tip"].mean()
print("Average Tip by Day:")
print(avg_tip_by_day)
```

```python
avg_tip_by_day.plot(kind="bar", title="Average Tip by Day", ylabel="Average Tip ($)", xlabel="Day of Week")
plt.show()
```

```python
total_bill_by_time = tips.groupby("time")["total_bill"].sum()
print("Total Bill by Time:")
print(total_bill_by_time)

total_bill_by_time.plot(kind="bar", title="Total Bill by Time", ylabel="Total Bill ($)", xlabel="Meal Time")
plt.show()
```

```python
tips["tip_percent"] = 100 * tips["tip"] / tips["total_bill"]
```

```python
avg_tip_percent_by_smoker = tips.groupby("smoker")["tip_percent"].mean()
print("Average Tip % by Smoker:")
print(avg_tip_percent_by_smoker)

avg_tip_percent_by_smoker.plot(kind="bar", title="Average Tip % by Smoker", ylabel="Tip %", xlabel="Smoker")
plt.show()
```

```python
# Save to CSV
avg_tip_percent_by_smoker.to_csv("results/avg_tip_percent_by_smoker.csv")
print("âœ… Saved avg_tip_percent_by_smoker.csv")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set a Seaborn style
sns.set(style="whitegrid")

# Bar chart: Average tip by day
plt.figure(figsize=(8, 5))
sns.barplot(x="day", y="tip", data=df)
plt.title("Average Tip by Day")
plt.ylabel("Tip ($)")
plt.xlabel("Day of Week")
plt.tight_layout()
plt.savefig("results/avg_tip_by_day_bar.png")
plt.show()
```

```python
# Ensure tip_percent is calculated
df["tip_percent"] = df["tip"] / df["total_bill"] * 100

# Box plot: Tip percent by smoker
plt.figure(figsize=(8, 5))
sns.boxplot(x="smoker", y="tip_percent", data=df)
plt.title("Tip Percent by Smoker Status")
plt.ylabel("Tip Percentage")
plt.xlabel("Smoker (Yes or No)")
plt.tight_layout()
plt.savefig("results/tip_percent_by_smoker_box.png")
plt.show()
```
